# 🎙️ Transcritor Mestre (Local & YouTube)
Este script transcreve e traduz áudios/vídeos locais ou links do YouTube utilizando o OpenAI Whisper.

```
# @title 🚀 Passo 2: Transcrever e Baixar TXT
import yt_dlp
import os
from google.colab import files
import glob

modo = "YouTube" # @param ["Arquivo_Local", "YouTube"]
url_youtube = "" # @param {type:"string"}
modelo_ia = "medium" # @param ["base", "small", "medium", "large-v3"]
traduzir_para_ingles = True # @param {type:"boolean"}

tarefa = "translate" if traduzir_para_ingles else "transcribe"
nome_final = "transcricao_resultado"

if modo == "YouTube":
    if not url_youtube:
        print("❌ Erro: Insira a URL do YouTube!")
    else:
        ydl_opts = {'format': 'm4a/bestaudio/best', 'outtmpl': f'{nome_final}.%(ext)s'}
        with yt_dlp.YoutubeDL(ydl_opts) as ydl: ydl.download([url_youtube])
        arquivo_processar = f"{nome_final}.m4a"
else:
    uploaded = files.upload()
    arquivo_processar = list(uploaded.keys())[0]

print(f"🤖 Processando {arquivo_processar} com modelo {modelo_ia}...")
!whisper "{arquivo_processar}" --model {modelo_ia} --task {tarefa} --output_format txt

nome_txt = arquivo_processar.rsplit('.', 1)[0] + ".txt"
if os.path.exists(nome_txt):
    files.download(nome_txt)
    print("✅ Download concluído!")
```
